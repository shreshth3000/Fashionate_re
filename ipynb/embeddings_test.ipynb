```python
from qdrant_client import QdrantClient
import os

qclient = QdrantClient(
    url=os.getenv('QDRANT_DB_URL'),
    api_key=os.getenv('QDRANT_API_KEY')
)

qclient
```

```python
from qdrant_client.http.models import Filter, ScrollRequest


collection_name = "images_and_clothes_swin"
all_points = []

# Paginate to collect all points
scroll_offset = None
while True:
    response = qclient.scroll(
        collection_name=collection_name,
        with_payload=True,
        with_vectors=False,  # change to True if you want old vectors
        limit=1000,
        offset=scroll_offset
    )
    points, scroll_offset = response
    if not points:
        break
    all_points.extend(points)
    if scroll_offset is None:
        break
```

```python
import pandas as pd

data = []

for point in all_points:
    row = {"id": point.id}
    row.update(point.payload)  # flatten payload
    data.append(row)

df = pd.DataFrame(data)
```
