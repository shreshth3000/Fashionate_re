```python
import os
import pandas as pd


df = pd.read_csv('../new_data.csv')
all_urls = df['img'].tolist()
all_urls[:10]
```

```python
base_directory_cloth = "clothes"
base_directory_image = "images"
all_cloth_urls = list(map(lambda item: f"{base_directory_cloth}/{item}", all_urls))
all_image_urls = list(map(lambda item: f"{base_directory_image}/{item}", all_urls))
all_cloth_urls[::10]
```

```python
df = df.rename(columns={'img': 'img_path'})
df['img_path'] = all_image_urls
df['cloth_path'] = all_cloth_urls
cols = df.columns.tolist()
img_path_index = cols.index('img_path')
cols.insert(img_path_index + 1, cols.pop(cols.index('cloth_path')))

df = df[cols]

desc_df = pd.read_csv('../description.csv')
df['description'] = desc_df['description']

df.replace("unknown", None, inplace=True)
payloads = df
payload_dicts = [
    {k: v for k, v in row.items() if pd.notnull(v)}
    for row in payloads.to_dict(orient="records")
]

```

```python
from PIL import Image
from pandas import DataFrame

local_directory_cloth = "../filtered_clothes"
local_directory_image = "../filtered_images"
local_cloth_urls = list(map(lambda item: f"{local_directory_cloth}/{item}", all_urls))
local_image_urls = list(map(lambda item: f"{local_directory_image}/{item}", all_urls))

url_df = DataFrame.from_records({"image_url":local_image_urls, "cloth_url":local_cloth_urls})
```

```python
images = [Image.open(path).copy() for path in url_df["image_url"]]
images
```

```python
clothes = [Image.open(path).copy() for path in url_df["cloth_url"]]
```

```python
from transformers import AutoFeatureExtractor, SwinModel
import torch


device = "cuda" if torch.cuda.is_available() else "cpu"
processor = AutoFeatureExtractor.from_pretrained("microsoft/swin-base-patch4-window7-224")
model = SwinModel.from_pretrained("microsoft/swin-base-patch4-window7-224").to(device)

inputs = processor(
    list(images),
    return_tensors = "pt",
)

inputs = {k: v.to(device) for k, v in inputs.items()}

outputs = model(**inputs)
embeddings = outputs.last_hidden_state(dim=1)

embeddings
```

```python
from transformers import AutoFeatureExtractor, SwinModel
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
processor = AutoFeatureExtractor.from_pretrained("microsoft/swin-base-patch4-window7-224")
model = SwinModel.from_pretrained("microsoft/swin-base-patch4-window7-224").to(device)

batch_size = 16  # You can adjust this based on your GPU memory
embeddings = []

for i in range(0, len(clothes), batch_size):
    batch = clothes[i:i + batch_size]

    inputs = processor(
        list(batch),
        return_tensors="pt",
        padding=True
    )
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)
        batch_embeddings = outputs.last_hidden_state.mean(dim=1)  # (batch_size, 1024)
        embeddings.append(batch_embeddings.cpu())  # Move to CPU to avoid GPU memory overload

embeddings = torch.cat(embeddings, dim=0)
```

```python
from qdrant_client.models import Filter, FieldCondition, MatchValue


collection_name = "images_and_clothes_swin"

# 1) Build a filter for bought == True
bought_filter = Filter(
    must=[ FieldCondition(key="bought", match=MatchValue(value=True)) ]
)

# 2) Scroll through all matching points (no payload/vector needed)
points, _ = qclient.scroll(
    collection_name=collection_name,
    scroll_filter=bought_filter,
    with_payload=False,
    with_vectors=False,
    limit=10_000  # adjust if you have more than 10k bought items; you can page otherwise
)

# 3) Extract their IDs
bought_ids = [p.id for p in points]

print("Bought point IDs:", bought_ids)
```

```python
from qdrant_client.http.models import Filter, FieldCondition, MatchValue

# 1. Define the filter to get all items where bought=True
bought_filter = Filter(
    must=[FieldCondition(key="bought", match=MatchValue(value=True))]
)

# 2. Scroll and get all matching point IDs
points, _ = qclient.scroll(
    collection_name=collection_name,
    scroll_filter=bought_filter,
    with_payload=False,
    with_vectors=False,
    limit=10000
)
point_ids = [p.id for p in points]

# 3. Set 'bought' = False for all those IDs
if point_ids:
    qclient.set_payload(
        collection_name=collection_name,
        payload={"bought": False},
        points=point_ids
    )
    print(f"Updated {len(point_ids)} items: bought â†’ False")
else:
    print("No items with bought=True found.")
```
